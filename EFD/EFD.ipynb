Ajuste bloco para EFD

```python
import io

# Lista para armazenar as linhas modificadas
linhas_modificadas = []

# Abrir o arquivo para leitura
with io.open('EFD L4B 05_2022.txt', 'r', encoding='ISO-8859-1') as file:
    # Ler todas as linhas do arquivo
    linhas = file.readlines()

# Processar cada linha do arquivo
for linha in linhas:
    if linha.startswith('|A100|'):
        # Se a linha começar com '|A100|', realizar a modificação
        partes = linha.strip().split('|')  # Dividir a linha pelo delimitador '|'
        
        # Verificar se o campo 12 está vazio ou zerado
        if partes[12] == '' or partes[12] == '0,00':
            # Substituir o valor faltante pela posição 15
            partes[12] = partes[15]
        
        # Reconstruir a linha modificada
        linha_modificada = '|'.join(partes) + '\n'
        linhas_modificadas.append(linha_modificada)
    else:
        # Caso contrário, adicionar a linha original
        linhas_modificadas.append(linha)

# Gravar todas as linhas de volta no arquivo, mantendo o conteúdo original e as modificações
with io.open('EFD L4B 05_2022_Ajustado.txt', 'w', encoding='ISO-8859-1') as file:
    file.writelines(linhas_modificadas)
```
