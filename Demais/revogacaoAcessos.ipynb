```python
import pandas as pd

base = pd.read_excel('Loggi_Issue_Tracker.xlsx', sheet_name='Loggi Issue Tracker')
perfis_sap = pd.read_excel('Loggi_Issue_Tracker.xlsx', sheet_name='Perfis SAP')
perfis_pipefy = pd.read_excel('Loggi_Issue_Tracker.xlsx', sheet_name='Perfis Pipefy')

#Na coluna 'Descrição' preciso extrair apenas o texto que está entre dois * e colocar em uma nova coluna chamada 'e-mail' 
base['e-mail'] = base['Descrição'].str.extract(r'\*(.*?)\*')

#Deletar a coluna 'Descrição'
base = base.drop(columns=['Descrição'])

#Dentro da mesma planilha existem outras duas sheets chamadas 'Perfis SAP' e 'Perfis Pipefy' ambas tem uma coluna chamada 'E-mail'
#Preciso comparar a coluna 'e-mail' que criamos com as colunas das outras sheets e mostrar apenas os registros que estão  em uma das duas sheets de perfil.
base = base[base['e-mail'].isin(perfis_sap['e-mail']) | base['e-mail'].isin(perfis_pipefy['e-mail'])]

display(base)
```
